# PREDICTING IF CANCER IS MALIGNANT OR BENIGN

# Import the Dataset

```python
from sklearn.datasets import load_breast_cancer
```

```python
data=load_breast_cancer()
print(data)
```

# Splitting the dataset into features and target values

```python
features=data.data
targets=data.target
```

# Determine the shape of feature and target values

```python
print(features.shape)
print(targets.shape)
```

# Splitting of training and testing data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(features,targets,test_size=0.2)
```

# Shape of training and testing data

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Creating a neural network architecture


```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model=Sequential()
model.add(Dense(units=30,activation="relu",input_dim=30))
model.add(Dense(units=60,activation="relu"))
model.add(Dense(units=1,activation="sigmoid"))
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=['accuracy'])
model.summary()
```

# Training the data

```python
model.fit(x_train,y_train,epochs=500)
```

# Accuracy of our model

```python
model.evaluate(x_test,y_test)
```

# **ACCURACY ACHIEVED OF THE MODEL 97%**

# Predicting whether cancer is malignant or benign

```python
predictions = model.predict(x_test)
label = []
for prediction in predictions:
  if(prediction<0.5):
    label.append("BENIGN")
  else:
    label.append("MALIGNANT")

label
```
